```python
#!/usr/bin/env python3
import json
from dataclasses import dataclass
from typing import Dict, List

@dataclass
class Example:
    prompt: str
    response: str
    category: Dict[str, bool]
    is_safe: bool

def load_jsonl_dataset(path: str) -> List[Example]:
    examples: List[Example] = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            obj = json.loads(line)
            ex = Example(
                prompt=obj["prompt"],
                response=obj["response"],
                category=obj.get("category", {}),
                is_safe=bool(obj["is_safe"]),
            )
            examples.append(ex)
    return examples

if __name__ == "__main__":
    # Quick manual test
    path = "train.jsonl"  # change if your file name is different
    data = load_jsonl_dataset(path)
    print(f"Loaded {len(data)} examples.")
    for i, ex in enumerate(data[:5], start=1):
        print(f"\n=== Example {i} ===")
        print("Prompt:  ", ex.prompt)
        print("Response:", ex.response)
        print("is_safe: ", ex.is_safe)
```
